```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

# Varied seed and state, batch size fixed to 64

```python
data1 = np.loadtxt("results\diabetes-results-philox.txt", dtype=float)
data1_size = np.size(data1)

# Plot model performance
x_axis = [i for i in range(data1_size)]  # set x-axis to array indices
plt.figure(figsize=(8, 6), dpi=80)
plt.scatter(x_axis, data1, s=5)
plt.xlabel("Model")
plt.ylabel("Final accuracy")
plt.ylim([0,1])
plt.yticks(np.arange(0, 1, .1))
plt.title("ANN performance on PID data (batch_size=64)")
plt.savefig("ANN-PID-64.png", bbox_inches='tight')
plt.show()
```

```python
plt.close()
```

```python
max_accuracy = np.max(data1)
min_accuracy = np.min(data1)
print ("Max: ", max_accuracy)
print ("Min: ", min_accuracy)
print(max_accuracy - min_accuracy)
```
